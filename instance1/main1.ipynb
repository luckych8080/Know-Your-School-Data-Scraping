```python

# from xml.dom.expatbuilder import DOCUMENT_NODE
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
import time
# import json
# import pytesseract
# from PIL import Image
import csv


def scrape_school_info_modify(pincode):
    driver = webdriver.Chrome()

    driver.get('https://src.udiseplus.gov.in/')

    INPUT_PINCODE = 110053
    
    filename = "school_info_All.csv"


    # Create a CSV file and write the headers
    csv_file = open(filename, "w", newline="")
    writer = csv.writer(csv_file)
    headers = [
        "School Name",
        "Udise Code",
        "State",
        "District",
        "Block",
        "Cluster",
        "Village",
        "PINCODE",
        "School Category",
        "School Type",
        "Class To",
        "Class From",
        "State Management",
        "National Management",
        "Location",
        "blank",
        "I",
        "II",
        "III",
        "IV",
        "V",
        "VI",
        "VII",
        "VIII",
        "IX",
        "X",
        "XI",
        "XII",
        "Total 1-12",
        "Total with primary",
        "XXXX",
        "Teacher",
        "pta nhi"
    ]
    writer.writerow(headers)

    # time.sleep(10)

    # pin code click
    checkbox_element = driver.find_element(By.ID, "byPinCode")
    checkbox_element.click()
    # time.sleep(10)

    # Find the input element by its ID
    input_element = driver.find_element(By.ID, "search")

    # Clear any existing value in the input field
    input_element.clear()

    # Insert the desired value
    input_element.send_keys(INPUT_PINCODE)


    # Find the input element for the Captcha by its class name
    captcha_input = driver.find_element(By.CLASS_NAME, "capture")

    # Clear any existing value in the input field
    captcha_input.clear()

    # Prompt the user to enter the Captcha value manually
    captcha_value = input("Please enter the Captcha code: ")

    # Insert the Captcha value
    captcha_input.send_keys(captcha_value)

    # time.sleep(10)

    # Find the submit button element by its ID
    submit_button = driver.find_element(By.ID, "homeSearchBtn")

    # Perform a click action on the submit button
    actions = ActionChains(driver)
    actions.move_to_element(submit_button).click().perform()

    #
    # front page code end 
    #
    
    print(INPUT_PINCODE)

    for i in pincode:


        # pin code click
        checkbox_element = driver.find_element(By.ID, "byPinCodeOnSearchPage")
        checkbox_element.click()
        # time.sleep(10)

        # Find the input element by its ID
        input_element = driver.find_element(By.ID, "search")

        # Clear any existing value in the input field
        input_element.clear()

        # input_pincode = i
        # Insert the desired value
        input_element.send_keys(i)

        # time.sleep(10)

        # Find the submit button element by its ID
        submit_button = driver.find_element(By.CSS_SELECTOR, "div.go-icon button.btn.btn-info.btn-block")

        # Perform a click action on the submit button
        actions = ActionChains(driver)
        actions.move_to_element(submit_button).click().perform()

        # time.sleep(10)


        # Find the select element
        select_element = driver.find_element(By.NAME,"example_length")

        # Create a Select object
        select = Select(select_element)

        # Select the option with value "1000"
        select.select_by_value("1000")

        # Find the tbody element
        tbody = driver.find_element(By.XPATH, "//tbody")

        # Find all rows within the tbody
        rows = tbody.find_elements(By.XPATH, ".//tr")


        # Iterate through each row and extract the information
        for row in rows:
            # Extract the desired information from each cell within the row
            cells = row.find_elements(By.XPATH, ".//td")
        
            if len(cells) >= 6:

                pin_code = cells[5].text.strip().split(":")[-1].strip()

                # Find the link element
                if pin_code == '0-Operational':
                    # Find the link element within the current row
                    try:
                        link_element = row.find_element(By.CSS_SELECTOR, "td form a.clickLink")
                    except NoSuchElementException:
                    # Handle the exception (e.g., print a message or log it)
                        print("Link element not found. Continuing to the next element.")
                        
                        # Continue to the next iteration of the loop
                        continue
                    # Click on the link to open in a new tab
                    
                    link_element.click()

                    # Switch to the new tab
                    driver.switch_to.window(driver.window_handles[-1])

                    # Fetch the desired element data from the new tab
                    
                    # Wait for the card block element to be present
                    all_data = WebDriverWait(driver, 10).until(
                        EC.presence_of_element_located((By.CSS_SELECTOR, "div.card"))
                    )

                    my_array = []

                    # time.sleep(1)
                                
                    # # Extract the required information
                    card = all_data.find_element(By.CSS_SELECTOR, "div.card-block.p-3")

                    # Find all rows within the card-block div
                    rows = card.find_elements(By.CLASS_NAME, 'row')

                    # Iterate over the rows and extract the values
                    for row in rows:
                        cols = row.find_elements(By.CLASS_NAME,'col-md-6')
                        for col in cols:
                            label = col.find_element(By.TAG_NAME,'b').text
                            value = col.find_element(By.TAG_NAME, 'span').text
                            my_array.append(value.strip())

                    


                    # classes data 

                    # Find the div element to click
                    div_element = driver.find_element(By.CSS_SELECTOR, 'div.card-header.my-2')

                    # Click on the div element
                    ActionChains(driver).move_to_element(div_element).click().perform()

                    
                    # Find the first table element
                    tables = driver.find_elements(By.CLASS_NAME, 'table')
                    table_data = []

                    # Iterate over each table
                    for table in tables:
                        rows = table.find_elements(By.TAG_NAME, 'tr')
                        data = []

                        # Iterate over each row and extract the text from each cell
                        for row in rows:
                            cells = row.find_elements(By.TAG_NAME, 'td')

                            # Skip the header row
                            if not cells:
                                continue

                            row_data = [cell.text for cell in cells]
                            data.append(row_data)

                        table_data.append(data)

                    # Print the extracted table data
                    for data in table_data:
                        for row_data in data:
                            values = row_data[1:]  # Exclude the first element

                        # Print the second array value on each next line
                        for value in values:
                            my_array.append(value)

                        # Print the second value from the third array
                        if len(row_data) >= 3:
                            my_array.append(row_data[2])

                    
                    # print('---')

                    row_data = my_array
                    writer.writerow(row_data)

                
                    # Close the new tab
                    driver.close()

                    # Switch back to the previous tab
                    driver.switch_to.window(driver.window_handles[0])
        
                
                # Continue with any further actions or operations on the previous site

        print(i)




    # Close the CSV file and the webdriver
    csv_file.close()
    driver.quit()


```
